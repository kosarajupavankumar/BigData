```python
# prompt: Tell me about the string and from basic to adavce with more examples

# Basic String Operations

# Creating strings
my_string = "Hello, world!"
another_string = 'This is another string.'
multi_line_string = """This is a multi-line
string that can span multiple lines."""

# Accessing characters (indexing starts at 0)
print(my_string[0])  # Output: H
print(my_string[7])  # Output: w

# String length
print(len(my_string))  # Output: 13

# String concatenation
combined_string = my_string + " " + another_string
print(combined_string)

# String repetition
repeated_string = my_string * 3
print(repeated_string)

# String slicing
substring = my_string[0:5]  # Characters from index 0 up to (but not including) 5
print(substring)  # Output: Hello
substring2 = my_string[:5] # equivalent to my_string[0:5]
print(substring2) # Output: Hello
substring3 = my_string[7:]  # Characters from index 7 to the end
print(substring3) # Output: world!
substring4 = my_string[-6:] #Characters from 6th character from the end to the end of the string.
print(substring4) # Output: world!

# String methods

# Case conversion
print(my_string.lower())  # Output: hello, world!
print(my_string.upper())  # Output: HELLO, WORLD!

# Finding substrings
print(my_string.find("world"))  # Output: 7 (index of the first occurrence)
print(my_string.find("xyz")) # Output: -1 (if not found)

# Replacing substrings
new_string = my_string.replace("world", "Python")
print(new_string)  # Output: Hello, Python!


# Checking string properties
print(my_string.startswith("Hello"))  # Output: True
print(my_string.endswith("!"))  # Output: True
print(my_string.isalpha())  # Output: False (because of space and !)
print("hello".isalpha())  # Output: True
print(my_string.isdigit()) # Output: False
print("123".isdigit()) # Output: True
print(my_string.isalnum()) # Output: False


# Splitting and Joining
words = my_string.split(",") # Split string into a list of strings based on delimiter ','
print(words) #Output: ['Hello', ' world!']

joined_string = "-".join(words) # Join a list of strings into a single string with delimiter '-'
print(joined_string) #Output: Hello- world!

# Removing whitespace
string_with_spaces = "   Extra spaces   "
print(string_with_spaces.strip())  # Output: Extra spaces

# Advanced String Operations

# Formatting strings (f-strings)
name = "Alice"
age = 30
formatted_string = f"My name is {name} and I am {age} years old."
print(formatted_string)  # Output: My name is Alice and I am 30 years old.

# Regular expressions (import the 're' module)
import re

text = "My phone number is 123-456-7890."
match = re.search(r"\d{3}-\d{3}-\d{4}", text)
if match:
    phone_number = match.group(0)
    print(phone_number)  # Output: 123-456-7890

#String Encoding and Decoding
# encode converts string to bytes, decode converts bytes to string
string = "Hello, world!"
encoded_string = string.encode('utf-8') # Default encoding is utf-8. other encodings like latin-1, ascii can also be used
decoded_string = encoded_string.decode('utf-8')
print(encoded_string) #Output: b'Hello, world!'
print(decoded_string) #Output: Hello, world!


#Raw strings
raw_string = r"This is a raw string\n with no newline"
print(raw_string) # Output: This is a raw string\n with no newline

# Unicode
unicode_string = "你好，世界！"
print(unicode_string) # Output: 你好，世界！

```
